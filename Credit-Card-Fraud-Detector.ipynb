```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
data=pd.read_csv("/content/drive/My Drive/creditcard.csv")
data.head()
```

```python
pd.options.display.max_columns=None
```

```python
#All columns are displayed
data
```

```python
data.info()
```

```python
# No missing value
data.isnull().sum()
```

```python
#Collecting all numeric columns and checking if all values are scaled or not..
numeric_columns=data.select_dtypes(include='number')
isScaledall=numeric_columns.apply(
    lambda col: col.mean()==0 and col.std()==1
)
isScaledall.all()
```

```python
#hence its returning False , it means data needs proper scaling..
#visualizing unscaled data...
numeric_columns.boxplot(vert=False,figsize=(12,8))
plt.title="Box plot of numeric columns"
plt.show()
```

```python
# we can see 'Amount' column has unscaled data..
```

```python
# data-preprocessing...
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data['Amount']=sc.fit_transform(data['Amount'].to_numpy().reshape(-1,1))
data.head()
```

```python
data.select_dtypes(include='number').boxplot(vert=False,figsize=(12,8))
plt.title("Box plot of numeric columns")
plt.show()
```

```python
# data scaling has done properly now..
```

```python
#checking for duplicates..
data.duplicated().sum()
```

```python
#removal of all duplicates...
data=data.drop_duplicates()
data.duplicated().sum()
```

```python
data['Class'].value_counts()
```

```python
normal=data[data['Class']==0]
fraud=data[data['Class']==1]
print(f"Normal Transactions: {normal.shape[0]}\nFraudulent Transactions: {fraud.shape[0]}")
```

```python
# Dataset is highly imbalanced
# visualization...
#sns.countplot(data=data['Class'],color='purple')
#plt.xlabel="Class"
#plt.show()
```

```python
# undersampling is used first to handle imbalanced dataset..
```

```python
normal_sample=normal.sample(n=fraud.shape[0])
normal_sample.shape
```

```python
fraud.shape
```

```python
#both data is having same dimension now after applying undersampling...
#new data set...
new_data=pd.concat([normal_sample,fraud],axis=0,ignore_index=True)
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
# Splitting dataset into training data and testing data...
X=new_data.drop('Class',axis=1) # X is feature variable which is taking independent dataset
y=new_data['Class']             # y is respose variable which is taking dependent dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
# X_train and y_train will be provided for model training
# X_test will be given for performing predictions
# y_test will be using for comparision with the predicted data or, for measuring model performance
```

```python
# Model 1: Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred1=lr.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_pred2=dt.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred3=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
#Calculating perfoances of each three model and storing into a dataframe..

performance=pd.DataFrame({
    'Models':['LR','DTC','RFC'],
    'Accuracy_score':[accuracy_score(y_test,y_pred1)*100,
                      accuracy_score(y_test,y_pred2)*100,
                      accuracy_score(y_test,y_pred3)*100],
    'Precision_score':[precision_score(y_test,y_pred1)*100,
                       precision_score(y_test,y_pred2)*100,
                       precision_score(y_test,y_pred3)*100],
    'Recall_score':[recall_score(y_test,y_pred1)*100,
                    recall_score(y_test,y_pred2)*100,
                    recall_score(y_test,y_pred3)*100],
    'f1_score':[f1_score(y_test,y_pred1)*100,
                f1_score(y_test,y_pred2)*100,
                f1_score(y_test,y_pred3)*100]
})
performance
```

```python
sns.barplot(performance)
```

```python
# melt the dataframe for better visualization
melted_data=pd.melt(performance, id_vars='Models', var_name='Metric', value_name='Score')
# Set the style of seaborn
sns.set(style="whitegrid")
# create a bar plot
plt.figure(figsize=(12,8))
sns.barplot(x='Models',y='Score',data=melted_data,palette='viridis')
#set the title and labels
plt.xlabel="Models"
plt.ylabel="Score"
plt.title="Model metrics comparison"
#show the plot
plt.show()
```

```python
# Hence its showing LOGISTIC REGRESSION IS THE BEST MODEL IF WE APPLY UNDERSAMPLING..
```

```python
# Oversampling method is applied next for handling unbalanced dataset 'data'
from imblearn.over_sampling import SMOTE
X=data.drop('Class',axis=1)
y=data['Class']
X_res,y_res=SMOTE().fit_resample(X,y)
y_res.value_counts()
```

```python
# dataet is balanced by oversampling
```

```python
#splitting the balanced dataset into training data and testing data..
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.20,random_state=42)
```

```python
# Model 1: Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred1=lr.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_pred2=dt.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred3=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
#Calculating perfoances of each three model and storing into a dataframe..

final_data=pd.DataFrame({
    'Models':['LR','DTC','RFC'],
    'Accuracy_score':[accuracy_score(y_test,y_pred1)*100,
                      accuracy_score(y_test,y_pred2)*100,
                      accuracy_score(y_test,y_pred3)*100],
    'Precision_score':[precision_score(y_test,y_pred1)*100,
                       precision_score(y_test,y_pred2)*100,
                       precision_score(y_test,y_pred3)*100],
    'Recall_score':[recall_score(y_test,y_pred1)*100,
                    recall_score(y_test,y_pred2)*100,
                    recall_score(y_test,y_pred3)*100],
    'f1_score':[f1_score(y_test,y_pred1)*100,
                f1_score(y_test,y_pred2)*100,
                f1_score(y_test,y_pred3)*100]
})
final_data
```

```python
# melt the dataset for better visualization..
melted_data=pd.melt(final_data,id_vars='Models',var_name='Metric',value_name='Score')
# Set the style of seaborn
sns.set(style="whitegrid")
# create a barplot
plt.figure(figsize=(12,8))
sns.barplot(x='Models',y='Score',data=melted_data,palette='viridis')
#set the title and labels
plt.xlabel="Models"
plt.ylabel="Score"
plt.title="Model metrics comparison"
#show the plot
plt.show()
```
