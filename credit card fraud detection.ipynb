# Undersampling...

```python
normal=data[data['Class']==0]
fraud=data[data['Class']==1]
```

```python
normal.shape
```

```python
fraud.shape
```

```python
normal_sample=normal.sample(n=fraud.shape[0])
new_data=pd.concat([normal_sample,fraud],ignore_index=True)
new_data['Class'].value_counts()
```

```python
new_data.head()
```

```python
X=new_data.drop('Class',axis=1)  # feature dataframe --> independent
y=new_data['Class']             # Response series --> dependent
```

```python
#Split the dataset into Training Set and Data Set...
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred1=log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_pred2=dt.predict(X_test)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred3=rf.predict(X_test)
```

## storing info into a dataframe

```python
final_data=pd.DataFrame({
    'Models':['LR','DT','RF'],
    'Accuracy_score': [accuracy_score(y_test,y_pred1)*100,
            accuracy_score(y_test,y_pred2)*100,
            accuracy_score(y_test,y_pred3)*100,
           ],
       'Precision_score': [precision_score(y_test,y_pred1)*100,
            precision_score(y_test,y_pred2)*100,
            precision_score(y_test,y_pred3)*100,
           ],
     'Recall_Score': [recall_score(y_test,y_pred1)*100,
            recall_score(y_test,y_pred2)*100,
            recall_score(y_test,y_pred3)*100,
           ],
      'f1_score': [f1_score(y_test,y_pred1)*100,
            f1_score(y_test,y_pred2)*100,
            f1_score(y_test,y_pred3)*100,
           ] 
      })
final_data
```

```python
#Visualization....
sns.barplot(final_data.drop('Models',axis=1))
plt.show()
plt.xlabel="Models"
plt.ylabel="Accuracy"
```

```python
# Melt the dataframe for better visualization
melted_data = pd.melt(final_data, id_vars='Models', var_name='Metric', value_name='Score')

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Models', y='Score', data=melted_data, palette="viridis")
# Set the title and labels
plt.title('Model Metrics Comparison')
plt.xlabel='Models'
plt.ylabel='Score'

# Show the plot
plt.show()
```
